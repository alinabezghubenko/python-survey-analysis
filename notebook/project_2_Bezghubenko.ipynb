Завдання 3. Статистичний аналіз досвіду респондентів (WorkExp)

```python
cleaned_work_exp=df['WorkExp'].dropna() # очищаю колонку від пропущених значень

mean_val = round(cleaned_work_exp.mean(), 2)
median_val = cleaned_work_exp.median()
mode_val = cleaned_work_exp.mode()[0]  # витягую перше значення з результату mode

work_exp_stats = pd.Series({
    'Mean': mean_val,
    'Median': median_val,
    'Mode': mode_val})

work_exp_stats
```

Завдання 7. Географічний аналіз компенсації Python-розробників

Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
age_total = df.groupby('Age').size()

python_respondents = df[python_users_mask2]

age_python = python_respondents.groupby('Age').size()

age_python_pct = (age_python / age_total * 100).fillna(0) #.fillna(0) на випадок, якщо в віковій категорії не виявиться Python-розробників

age_analysis_df = age_python_pct.reset_index()
age_analysis_df.columns = ['Вікова категорія', 'Відсоток Python-розробників']

age_analysis_df['Відсоток Python-розробників'] = age_analysis_df['Відсоток Python-розробників'].round(1)
age_analysis_df
```

Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
salary_threshold = df['ConvertedCompYearly'].quantile(0.75) #поріг, вище якого знаходяться 25% найбільш оплачуваних розробників

high_paid_remote_mask = (df['ConvertedCompYearly'] > salary_threshold) & \
                        (df['RemoteWork'].str.contains('Remote', na=False))

remote_pros_df = df[high_paid_remote_mask].copy()

industry_stats = remote_pros_df['Industry'].value_counts() # .value_counts() автоматично сортує за частотою (від найбільш до найменш поширених)

top_industries = industry_stats.reset_index()
top_industries.columns = ['Індустрія', 'Кількість спеціалістів']

top_industries.head(10)
```
